```python
import pandas as pd

# This tells Python to read the file you downloaded
df = pd.read_csv('my_products.csv')

# This prints out the names of the columns
print(df.columns.tolist())
```

```python
# Make all the text lowercase so we don't miss anything 
df['Ingredients'] = df['Ingredients'].astype(str).str.lower()

# Create new columns that check if a specific word is inside the ingredient list (1 for yes, 0 for no)
df['has_salicylic'] = df['Ingredients'].str.contains('salicylic').astype(int)
df['has_niacinamide'] = df['Ingredients'].str.contains('niacinamide').astype(int)
df['has_hyaluronic'] = df['Ingredients'].str.contains('hyaluronic').astype(int)
df['has_ceramide'] = df['Ingredients'].str.contains('ceramide').astype(int)

# Let's check our work! Print out the product name and our new number columns
print("Data cleaned! Here is a sneak peek at your new features:")
print(df[['Name', 'has_salicylic', 'has_hyaluronic']].head(20))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# Pick our clues (the ingredients we found)
features = ['has_salicylic', 'has_niacinamide', 'has_hyaluronic', 'has_ceramide']
X = df[features]

# Pick our answer (the skin type it treats)
y = df['Target_Skin']

# Train the model to map the clues to the answer
model = DecisionTreeClassifier(random_state=42)
model.fit(X, y) # training means the model looks at all the clues and learns how they relate to the skin type

print("Brain successfully trained! The model now understands your inventory.")
```

```python
def my_shop_recommender():
    print("\n✨ WELCOME TO THE HER HAPPY SKIN MATCHER ✨")
    
    # Ask the user for input
    print("What is your skin type?")
    user_skin = input("Type exactly: Oily, Dry, Sensitive, or Combination: ").strip()
    
    # Filter your inventory for the exact match
    matches = df[df['Target_Skin'].str.lower() == user_skin.lower()]
    
    # Print out the results
    if len(matches) == 0:
        print(f"\nHmm, we might be out of stock for {user_skin} skin right now.")
    else:
        print(f"\nFound {len(matches)} perfect products for {user_skin} skin in your stock:\n")
        
        for index, row in matches.iterrows():
            print(f"🛍️ Product: {row['Name']}")
            print(f"   Contains: {row['Ingredients']}")
            print("-" * 40)

# Run your new app!
my_shop_recommender()
```
